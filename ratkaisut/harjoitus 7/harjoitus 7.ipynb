Liitetiedostossa on kondensaattorin lautausj√§nnite U (V) ja kondensaattorin lataus Q (Coulombi) arvoja. Esit√§ mittaustulokset (j√§nnite, varaus)-kuvaajassa, ja m√§√§rit√§ kuvaajan kulmakertoimesta  kondensaattorin kapasitanssi C.  Kondensaattorilaki Q= CU

```python
with open("kondensaattorilaki.csv") as f:
    rivit = f.read().splitlines()
    
U_lista = []
Q_lista = []
for rivi in rivit:
    U, Q = rivi.split(",")
    U = float(U)
    Q = float(Q)
    
    U_lista.append(U)
    Q_lista.append(Q)
    
kuvaaja.plot(U_lista, Q_lista)
kuvaaja.grid()
kuvaaja.xlabel("U (j√§nnite)")
kuvaaja.ylabel("Q (varaus)")
kuvaaja.show()    

deltaQ = Q_lista[-1] - Q_lista[0]
deltaU = U_lista[-1] - U_lista[0]

print(deltaQ)
print(deltaU)
k = deltaQ/deltaU

print(k)
```

Tasaisesti kiihtyv√§ liike: Piirr√§ liikkuvan kappaleen (aika, sijainti)-kuvaaja sek√§ (aika, nopeus)-kuvaaja. M√§√§rit√§ nopeuden kuvaajasta graafisen integroinnin avulla kuljettu matka.

```python
with open("tasaisesti kiihtyva liike.csv") as f:
    rivit = f.read().splitlines()
#print(rivit)

t_lista = []
s_lista = []
v_lista = []
a_lista = []
for rivi in rivit:
    t, s, v, a = rivi.split(",")
    t = float(t)
    s = float(s)
    v = float(v)
    a = float(a)
    
    t_lista.append(t)
    s_lista.append(s)
    v_lista.append(v)
    a_lista.append(a)

    
kuvaaja.plot(t_lista, s_lista)
kuvaaja.grid()
kuvaaja.show()

kuvaaja.plot(t_lista, v_lista)
kuvaaja.grid()
kuvaaja.show()

kuvaaja.plot(t_lista, a_lista)
kuvaaja.grid()
kuvaaja.show()

i = 0
while v_lista[i] > 0:
    i = i + 1
deltat = t_lista[i]
print(deltat)

kohta = 0
suurin_kohta = 0
suurin_s = s_lista[0]
for s in s_lista:
    if s > suurin_s:
        suurin_kohta = kohta
        suurin_s = s
    kohta = kohta + 1

print(v_lista[suurin_kohta])
```
